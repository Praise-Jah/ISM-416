# Article One

```python
import requests
from bs4 import BeautifulSoup
from gensim.summarization import summarize
```

```python
import requests
from bs4 import BeautifulSoup
import gensim
import docx
url = 'https://www.bbc.com/news/world-asia-china-57267729'
page = requests.get(url).text
soup = BeautifulSoup(page)
header = soup.find('h1').get_text()
header2 = 'Summary'
time = soup.find('time', attrs={'data-testid':'timestamp'}).get_text()
time
p_tags = soup.find_all('p')
text = [tags.get_text() for tags in p_tags]
print(text)
sentence_list = [sentence for sentence in text if not '\n' in sentence]
sentence_list = [sentence for sentence in sentence_list if '.' in sentence]
article = ''.join(sentence_list)
from gensim.summarization import summarize
summary = summarize(article, ratio = 0.30)
doc = docx.Document()
doc.add_heading(header)
doc.add_picture('img1.png')
doc.add_paragraph(time)
doc.add_paragraph(article)
doc.add_heading(header2)
doc.add_paragraph(summary)
doc.save('NlpTasksassignment1')
```

## Tokenization

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(summary)
```

```python
for token in doc:
    print(token.text, '--->', spacy.explain(token.text))
```

```python
for token in doc:
    print(token.text, '--->', token.pos_, '--->', spacy.explain(token.pos_))
```

```python
for token in doc:
    print(token.text, '--->', token.dep_, '--->', spacy.explain(token.dep_))
```

```python
for token in doc:
    print(token.text, '--->', token.lemma_, '--->', spacy.explain(token.lemma_))
```

## Sentence boundary detection

```python
list(doc.sents)
```

```python
sentences = list(doc.sents)
```

```python
for sentence in sentences:
    print(sentence)
```

## Named Entity

```python
for ent in doc.ents:
    print (ent.text, '--->', ent.label_, '--->', spacy.explain(ent.label_))
```

## Visualization

```python
from spacy import displacy
```

```python
displacy.render(doc, style = 'ent', jupyter = True)
```

```python
displacy.render(doc, style = 'dep', jupyter = True)
```
